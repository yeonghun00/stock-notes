```python
import yfinance as yf

symbol = '^GSPC'
data = yf.Ticker(symbol)
df = data.history(period='1m', start='1985-1-1', end='2022-7-1')
```

```python
year_df = df.resample('Y').first()
year_df
```

```python
import numpy as np
```

```python
m = np.mean(year_df['Close'][1:].values / year_df['Close'][:-1].values)
var = np.var(year_df['Close'][1:].values / year_df['Close'][:-1].values)
```

```python
print(m)
print(var)
```

```python
import math
import random

arr = [random.gauss(m, math.sqrt(var)) for i in range(50)]
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(20,10))
plt.plot(np.cumprod(arr))
plt.show()
```
