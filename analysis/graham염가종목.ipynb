```python
# https://opendart.fss.or.kr/disclosureinfo/fnltt/dwld/main.do
# ifrs-full_CurrentAssets	- ifrs-full_Liabilities
```

```python
from google.colab import files

# Prompt the user to upload a file
uploaded = files.upload()

# Access the uploaded file
for filename in uploaded.keys():
    print(f"Uploaded file: {filename}")
```

```python
file_content = uploaded[filename].decode('utf-8')
```

```python
file_content
```

```python
import pandas as pd
from io import StringIO  # 파일 내용을 읽어오기 위한 StringIO 사용

df = pd.read_csv(StringIO(file_content))
```

```python
df = df.pivot_table(index=['종목코드', '회사명', '보고서종류'], columns='항목코드', values='당기 1분기말', aggfunc='first').reset_index()
```

```python
df = df.drop(columns=['ifrs-full_Assets'])
```

```python
df['ifrs-full_Liabilities'] = df['ifrs-full_Liabilities'].fillna('0').str.replace(',', '').astype(int)
```

```python
df
```

```python
df = df.dropna()
```

```python
df['ifrs-full_CurrentAssets'] = df['ifrs-full_CurrentAssets'].apply(str).str.replace(',', '').astype(int)
df['ifrs-full_Liabilities'] = df['ifrs-full_Liabilities'].apply(str).str.replace(',', '').astype(int)
df
```

```python
df['순유동자산'] = df['ifrs-full_CurrentAssets'] - df['ifrs-full_Liabilities']
```
